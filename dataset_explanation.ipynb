## Credit Score calculation

Credit scores are numerical representations of a customer's creditworthiness, calculated based on their credit history, payment behavior, and financial habits. Different models exist for calculating credit scores, each using slightly different methodologies. Below are some common credit score definitions and calculation methods:

### **1. FICO Score (Fair Isaac Corporation)**
   - One of the most widely used credit scores, ranging from **300 to 850**.
   - Calculation is based on the following factors:
     - **Payment History (35%)** – On-time vs. late payments.
     - **Amounts Owed (30%)** – Credit utilization ratio.
     - **Length of Credit History (15%)** – Time since accounts were opened.
     - **Credit Mix (10%)** – Types of credit accounts (loans, credit cards, etc.).
     - **New Credit (10%)** – Number of recently opened accounts and inquiries.

### **2. VantageScore (Developed by Equifax, Experian, and TransUnion)**
   - Similar to FICO, ranges from **300 to 850**.
   - Uses a slightly different weighting:
     - **Total Credit Usage & Balances (Extremely Influential)**
     - **Credit Mix & Experience (Highly Influential)**
     - **Payment History (Moderately Influential)**
     - **Age of Credit (Less Influential)**
     - **New Accounts & Credit Inquiries (Less Influential)**

### **3. Equifax Credit Score**
   - Ranges from **280 to 850**.
   - Factors include:
     - **On-time payments**
     - **Credit utilization**
     - **Credit age**
     - **New credit inquiries**
     - **Types of credit used**

### **4. Experian PLUS Score**
   - A consumer educational score, **not used by lenders**.
   - Uses a similar scale and methodology as FICO but does not directly impact lending decisions.

### **5. TransUnion Credit Score**
   - Based on proprietary algorithms similar to FICO and VantageScore.
   - Often used in **free credit reports** or monitoring services.

### **6. Bank-Specific Credit Scoring Models**
   - Many banks and lenders develop their own custom credit scoring models based on their risk assessment strategies.
   - These models may integrate:
     - Banking transaction history.
     - Income and employment status.
     - Debt-to-income ratio.
     - Loan default probability.

### **7. AI & Machine Learning-Based Credit Scoring**
   - Uses deep learning models, federated learning, or adversarial debiasing techniques to predict creditworthiness.
   - Features may include:
     - Alternative credit data (e.g., rental payments, utility bills).
     - Behavioral data (e.g., online shopping habits, mobile payment history).
     - Explainability & bias mitigation techniques.

### **8. Credit Score Calculation Formula (Simplified)**
